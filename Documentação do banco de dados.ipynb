#### Sobre o conjunto de dados
##### Contexto:
Esta publicação de dados contém um banco de dados espacial de incêndios florestais que ocorreram nos Estados Unidos de 1992 a 2015. É a terceira atualização de uma publicação originalmente gerada para apoiar o sistema nacional de Análise de Programa de Incêndio (FPA). Os registros de incêndios florestais foram adquiridos dos sistemas de relatórios de organizações federais, estaduais e locais de incêndio. Os seguintes elementos de dados principais foram necessários para que os registros fossem incluídos nesta publicação de dados: data da descoberta, tamanho final do incêndio e uma localização do ponto pelo menos tão precisa quanto a seção do Sistema de Levantamento de Terras Públicas (PLSS) (grade de 1 milha quadrada). Os dados foram transformados para se adequar, quando possível, aos padrões de dados do National Wildfire Coordinating Group (NWCG). A verificação básica de erros foi realizada e os registros redundantes foram identificados e removidos, na medida do possível. O produto resultante,

- DISCOVERY_TIME =  Hora do dia em que o incêndio foi descoberto ou confirmado.

- STATCAUSEDESCR = Descrição da causa (estatística) do incêndio.
- CONT_DATE = Data em que o incêndio foi declarado contido ou controlado de outra forma (mm/dd/aaaa onde mm=mês, dd=dia e aaaa=ano).
- CONT_DOY = Dia do ano em que o fogo foi declarado contido ou controlado.
- CONT_TIME = Hora do dia em que o incêndio foi declarado contido ou controlado de outra forma (hhmm onde hh=hora, mm=minutos).

- LONGITUDE = Longitude (NAD83) para localização do ponto do incêndio (graus decimais).
- OWNER_CODE = Código do proprietário principal ou entidade responsável pela gestão do terreno no ponto de origem do incêndio no momento do incidente.
- OWNER_DESCR = Nome do proprietário principal ou entidade responsável pela gestão do terreno no ponto de origem do incêndio no momento do incidente.
- STATE = Código alfabético de duas letras para o estado em que o fogo queimou (ou se originou), com base na designação nominal no relatório de incêndio.
- COUNTY = Condado, ou equivalente, em que o fogo queimou (ou se originou), com base na designação nominal no relatório de incêndio.
- FIPS_CODE = Código de três dígitos da publicação 6-4 do Federal Information Process Standards (FIPS) para representação de condados e entidades equivalentes.
- FIPS_NAME = Nome do município da publicação FIPS 6-4 para representação de municípios e entidades equivalentes.
